```python
import cv2
import numpy as np
```

READ A IMAGE

```python
img=cv2.imread("bird.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
flat=img.flatten()
```

```python
msg = (
    "A gaurav is a distinct section of a written text, typically consisting of several sentences, "
    "that focuses on a single topic or idea. It helps organize and structure a piece of writing "
    "by breaking it into manageable, coherent units for the reader. Effective paragraphs usually "
    "start with a topic sentence that states the main point, followed by supporting sentences "
    "that provide details, evidence, and examples."
) + chr(255)
```

```python
binary=''.join(format(ord(ch),'08b')for ch in msg)
```

```python
binary
```

ENCODE A IMAGE

```python
if len(binary)>len(flat):
    raise SystemExit("image is too small")
```

```python
for i in range(len(binary)):
    flat[i]=(flat[i]& 0xFE) | int(binary[i])
```

```python
encoded=flat.reshape(img.shape)
cv2.imwrite("secret_bird.png",encoded)
print("message is hidden")
```

DECODED A IMAGE

```python
encode_img=cv2.imread("secret_bird.png",cv2.IMREAD_GRAYSCALE)
flat_dec=encode_img.flatten()
```

```python
bits= [str(flat_dec[i] & 1)for i in range(len(flat_dec))]
binary_str=''.join(bits)
```

```python
decoded_msg=" "
for i in range(0,len(binary_str), 8):
    byte = binary_str[i:i+8]
    if byte=='11111111':
        break
    decoded_msg +=chr(int(byte ,2))
```
